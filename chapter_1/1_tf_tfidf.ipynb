# 1. TF 표현 (Term Frequency, 문서 빈도)
- TF(w) : 단어 w의 문서(문장) 내 단순 빈도
- --> 단어 가방 모델(BoW)과 연관 
## 원-핫 벡터 만들기

```python
from sklearn.feature_extraction.text import CountVectorizer
import seaborn as sns

corpus = ["Time flies like an arrow.",\
         "Fruit flies like a banana."]
one_hot_vectorizer = CountVectorizer(binary=True)
one_hot = one_hot_vectorizer.fit_transform(corpus).toarray()
vocab = one_hot_vectorizer.get_feature_names()
sns.heatmap(one_hot, annot=True,
           cbar=False, xticklabels=vocab,
           yticklabels=['Sentence 1', 'Sentence 2'])
```

# 2. TF-IDF 표현 (Inverse Doc Freq, 문서-역문서 빈도)
- TF-IDF(w): TF(w)*IDF(w) = TF(w)*log(N/n)
- N: 전체 문서 수, n: 해당 단어 w가 포함된 문서 수
- 의미: 다른 문서에서도 흔한 단어에 패널티 부여, 특정 문서에만 존재하는 단어에 대한 중요도 높임.
- ex1) 모든 문서에 등장: IDF(W)=log1=0 --> TF-IDF(w)=0 --> 이 단어를 완전히 제외함.
- ex2) 한 문서에만 등장: IDF(w)=logN (최대값)
## TF-IDF 표현 만들기

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import seaborn as sns

tfidf_vectorizer = TfidfVectorizer()
tfidf = tfidf_vectorizer.fit_transform(corpus).toarray()
sns.heatmap(tfidf, annot=True, cbar=False, xticklabels=vocab,
           yticklabels=['Sentence 1', 'Sentence 2'])
```
